# **Introduction**

**Context and Problems**

Caffeine is one of the most widely consumed psychoactive substances globally, with millions relying on caffeinated beverages for energy and mood enhancement. While moderate caffeine intake can improve mental performance and reduce fatigue, excessive consumption has been linked to adverse effects such as poor sleep quality, increased stress, cardiovascular strain, and potential anxiety. These consequences manifest across multiple facets of well-being:

•	**Mental Well-being:** Elevated stress levels and impaired sleep quality diminish psychological health and cognitive function.

**•	Physical Well-being:** Cardiovascular effects such as increased heart rate and metabolic impacts reflected in BMI and health status indicate physical health risks.

**•	Lifestyle Factors:** Caffeine consumption patterns entangle with behaviours like smoking, alcohol use, physical activity, and demographic contexts, influencing overall well-being.

Understanding these effects is particularly important as caffeine consumption patterns vary across demographic groups, lifestyles, and occupational pressures. Many different populations, including students, healthcare workers, and high-stress professionals, may unknowingly expose themselves to health risks due to unregulated caffeine use. Moreover, widespread lack of awareness limits targeted preventive interventions and personalized guidance.


**Project Goal**

Leveraging a comprehensive 10,000-person dataset about demographic, caffeine intake, sleep patterns, stress levels, physiological measures, and lifestyle variables, this project aims to uncover how caffeine consumption influences mental, physical, and lifestyle facets of well-being across populations. The goal is to produce vital insights for individuals, health experts, workplaces, and businesses, supporting smarter consumption, reducing health risks, and fostering balanced lives in today’s fast-paced world.

**Hypothesis**

Research shows caffeine significantly impacts well-being across several dimensions. Sleep quality and duration deteriorate with caffeine intake, especially when consumption occurs near bedtime, leading to impaired cognitive and mental health (Sleep Foundation, 2025). Elevated stress may also result from excessive caffeine consumption, particularly in sensitive populations (Liu et al., 2024). Physical health risks include raised heart rate and potential cardiovascular strain (UC Davis Health, 2023) influenced variably by caffeine.

Key well-being metrics include:

**•	Positive:** Sleep quality, Sleep hours, Physical activity hours, BMI.

**•	Negative:** Stress level, Health issues, Heart rate, Smoking, Alcohol consumption.

**•	Caffeine-related:** Coffee intake, Caffeine (mg), shown to affect both positive and negative outcomes depending on dose and timing.

This project addresses gaps in integrating these various effects of caffeine, especially within diverse demographic and occupational contexts. Therefore, the project’s hypothesis is:

-> Higher caffeine intake associates with poorer sleep quality, elevated stress, and increased health risks involving cardiovascular and metabolic health, with variations shaped by age, occupation, lifestyle, and demographic factors.

```python
#Cleaning Text Data

print("Cleaning categorical columns:")
for col in cat_cols:
    if df[col].dtype == 'object': # Ensure it's a string column
        df[col] = df[col].str.strip() # Remove leading/trailing whitespace
        df[col] = df[col].str.lower() # Convert to lowercase
        df[col] = df[col].astype('category') # Convert to category dtype
        print(f"  - Cleaned and converted '{col}' to category dtype")

print("\nText data cleaning and category conversion complete. Displaying info and head to check:")
df.info()
df.head()
```
